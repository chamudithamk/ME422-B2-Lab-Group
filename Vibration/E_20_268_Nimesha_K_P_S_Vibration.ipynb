# Activity 1: Derivation of the 2-DOF Model

## 1. Physical Model and Governing Equation

The experimental apparatus is modeled as a continuous system using **Euler-Bernoulli Beam Theory**. For small amplitude transverse vibrations, the governing partial differential equation (PDE) is given by:

$$
\rho A \frac{\partial^2 y}{\partial t^2} + c \frac{\partial y}{\partial t} + EI \frac{\partial^4 y}{\partial x^4} = f(t, x)
$$

**Nomenclature:**
* $y(x,t)$: Transverse displacement at position $x$ and time $t$.
* $\rho A$: Mass per unit length.
* $EI$: Flexural rigidity.
* $c$: Distributed viscous damping coefficient.
* $f(t,x)$: External transverse force distribution.

### Boundary Conditions

The beam is pinned at the base ($x=0$) with a rotational spring stiffness $k_\theta$ (approximating the offset linear spring) and is free at the tip ($x=l_{beam}$).

**1. At x = 0 (Pinned with Elastic Constraint):**

* Zero Displacement condition:

$$
y(0, t) = 0
$$

* Moment Balance condition (Internal bending moment equals the restoring spring torque):

$$
EI \frac{\partial^2 y}{\partial x^2}(0, t) - k_\theta \frac{\partial y}{\partial x}(0, t) = 0
$$

**2. At x = l_beam (Free End):**

* Zero Bending Moment condition:

$$
\frac{\partial^2 y}{\partial x^2}(l_{beam}, t) = 0
$$

* Zero Shear Force condition:

$$
\frac{\partial^3 y}{\partial x^3}(l_{beam}, t) = 0
$$

---

## 2. Modal Analysis (Free Vibration)

To solve the eigenvalue problem, free vibration is assumed ($f=0, c=0$) with a separable solution $y(x,t) = \psi(x)e^{i\omega t}$. This reduces the PDE to the spatial ordinary differential equation:

$$
\frac{d^4\psi}{dx^4} - \beta^4\psi = 0
$$

where the wavenumber $\beta$ is related to the natural frequency $\omega$ by $\beta^4 = \frac{\rho A \omega^2}{EI}$. The general solution is:

$$
\psi(x) = A \cosh(\beta x) + B \sinh(\beta x) + C \cos(\beta x) + D \sin(\beta x)
$$

Substitution of the boundary conditions into the general solution yields a homogeneous system of linear equations for coefficients $A, B, C, D$. Specifically, the moment condition at $x=0$ requires:

$$
2EI \beta^2 A - k_\theta \beta (B + D) = 0
$$

The characteristic equation is obtained by setting the determinant of the coefficient matrix to zero. Numerical solution of this equation yields the wavenumbers $\beta_k$ and the corresponding mode shapes $\psi_k(x)$.

---

## 3. Discretization via Assumed Modes

The continuous system is discretized to a 2-Degree-of-Freedom (2-DOF) model using the **Assumed Modes Method**. The total displacement is approximated as a linear combination of the first two dominant mode shapes:

$$
y(t, x) \approx z_1(t)\psi_1(x) + z_2(t)\psi_2(x)
$$

where $z_1(t)$ and $z_2(t)$ are the generalized time-dependent coordinates. Substituting this expansion into the governing PDE and applying orthogonality properties yields the matrix equation of motion:

$$
\mathbf{M}\ddot{\mathbf{z}} + \mathbf{C}\dot{\mathbf{z}} + \mathbf{K}\mathbf{z} = \mathbf{F}(t)
$$

---

## 4. Derivation of System Matrices

The system matrices account for the distributed properties of the beam and the discrete contributions from the Spring, Damper, and Exciter attachments.

### A. Mass Matrix (M)

The total mass matrix is the sum of the distributed beam mass and the projected discrete masses.

$$
\mathbf{M} = \mathbf{M}_{beam} + \mathbf{M}_{spring} + \mathbf{M}_{damper} + \mathbf{M}_{exciter}
$$

**Beam Mass:** (Diagonal matrix due to orthogonality)

$$
M_{ii, beam} = \int_0^{l_{beam}} \rho A \psi_i^2(x) dx
$$

**Discrete Masses:**
For a discrete mass $m$ located at $x_a$, the contribution to the mass matrix is coupled:

$$
\mathbf{M}_{discrete} = m \begin{bmatrix} \psi_1^2(x_a) & \psi_1(x_a)\psi_2(x_a) \\ \psi_2(x_a)\psi_1(x_a) & \psi_2^2(x_a) \end{bmatrix}
$$

*Note: The spring mass contribution utilizes the effective mass derived from Rayleigh's Theory: $m_{eff} = \frac{1}{3}m_{spring} + m_{fixed}$.*

### B. Stiffness Matrix (K)

Since the potential energy of the rotational spring $k_\theta$ is intrinsic to the derived natural frequencies $\omega_k$, the stiffness matrix is diagonal:

$$
\mathbf{K} = \begin{bmatrix} \omega_1^2 M_{11,beam} & 0 \\ 0 & \omega_2^2 M_{22,beam} \end{bmatrix}
$$

### C. Damping Matrix (C)

The damping matrix includes structural modal damping and the discrete viscous damper.

$$
\mathbf{C} = \mathbf{C}_{beam} + \mathbf{C}_{damper}
$$

**Beam Damping:** Assumed modal damping with ratios $\zeta_1, \zeta_2$.

$$
\mathbf{C}_{beam} = \begin{bmatrix} 2\zeta_1 \omega_1 M_{11,beam} & 0 \\ 0 & 2\zeta_2 \omega_2 M_{22,beam} \end{bmatrix}
$$

**Viscous Damper:** (Located at $x_{damper}$ with coefficient $c_{damper}$)

$$
\mathbf{C}_{damper} = c_{damper} \begin{bmatrix} \psi_1^2(x_{damper}) & \psi_1(x_{damper})\psi_2(x_{damper}) \\ \psi_2(x_{damper})\psi_1(x_{damper}) & \psi_2^2(x_{damper}) \end{bmatrix}
$$

### D. Excitation Force Vector (F)

The harmonic excitation force $Q \sin(\Omega t)$ applied at $x_{exciter}$ is projected onto the generalized coordinates:

$$
\mathbf{F}(t) = \begin{bmatrix} Q \psi_1(x_{exciter}) \\ Q \psi_2(x_{exciter}) \end{bmatrix} \sin(\Omega t)
$$

---

## 5. Final Equations of Motion

The reduced-order model governing the generalized coordinates $\mathbf{z} = [z_1, z_2]^T$ is given by:

$$
\left( \mathbf{M}_{beam} + \sum \mathbf{M}_{discrete} \right) \ddot{\mathbf{z}} + \left( \mathbf{C}_{beam} + \mathbf{C}_{damper} \right) \dot{\mathbf{z}} + \mathbf{K}_{beam} \mathbf{z} = \mathbf{F}_{exciter}
$$
