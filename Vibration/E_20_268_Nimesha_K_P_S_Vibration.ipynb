# Activity #1
Derive the 2-dof freedom model that will capture the first two dominant modes of small amplitude vibration of the system.

---


## 1. Governing Equation

For small-amplitude transverse vibrations of a slender beam, the Euler--Bernoulli beam theory gives the governing partial differential equation as

$$\rho A \frac{\partial^2 y(x,t)}{\partial t^2} + c \frac{\partial y(x,t)}{\partial t} + EI \frac{\partial^4 y(x,t)}{\partial x^4} + P \frac{\partial^2 y(x,t)}{\partial x^2} = f(x,t)$$

where $\rho$ is the material density, $A$ is the cross-sectional area, $E$ is Young's modulus, $I$ is the second moment of area, $c$ is the distributed viscous damping coefficient, $P$ is the axial compressive force, and $f(x,t)$ is the external transverse load per unit length.

For the present case, axial loading is absent $(P=0)$, hence the governing equation reduces to

$$\rho A \frac{\partial^2 y}{\partial t^2} + c \frac{\partial y}{\partial t} + EI \frac{\partial^4 y}{\partial x^4} = f(x,t)$$

---

## 2. Boundary Conditions

The beam is elastically restrained in rotation at the base and free at the tip. Accordingly, the boundary conditions are

$$
\begin{aligned}
y(0,t) &= 0 \\
EI \frac{\partial^2 y}{\partial x^2}(0,t) - k_\theta \frac{\partial y}{\partial x}(0,t) &= 0 \\
\frac{\partial^2 y}{\partial x^2}(L,t) &= 0 \\
\frac{\partial^3 y}{\partial x^3}(L,t) &= 0
\end{aligned}
$$

where the rotational stiffness at the base is defined as

$$k_\theta = k_{\text{spring}} \, l_{\text{spring}}^2$$



---

## 3. Operator Formulation

Let $\mathcal{F}$ denote the space of four-times continuously differentiable functions defined on $[0,L]$. Define the linear operator $H : \mathcal{F} \rightarrow \mathcal{F}$ as

$$H = \frac{EI}{\rho A} \frac{\partial^4}{\partial x^4}$$

The equation of motion can then be written in operator form as

$$\frac{\partial^2 y}{\partial t^2} + C \frac{\partial y}{\partial t} + H y = q(x,t)$$

where $c = C \rho A$ and $f(x,t) = \rho A q(x,t)$.

---

## 4. Free Vibration and Eigenvalue Problem

Neglecting damping and external forcing, the free vibration problem is governed by

$$H \psi_k(x) = \omega_k^2 \psi_k(x)$$

Substituting the operator definition gives

$$\frac{EI}{\rho A} \frac{d^4 \psi_k}{dx^4} = \omega_k^2 \psi_k$$

or equivalently,

$$\frac{d^4 \psi_k}{dx^4} = \beta_k^4 \psi_k, \qquad \beta_k^4 = \frac{\rho A \omega_k^2}{EI}$$

---

## 5. General Solution

The general solution of the fourth-order differential equation is

$$\psi_k(x) = A_k \cosh(\beta_k x) + B_k \sinh(\beta_k x) + C_k \cos(\beta_k x) + D_k \sin(\beta_k x)$$

---

## 6. Application of Boundary Conditions

### 6.1 Base End $(x=0)$

From the displacement boundary condition,

$$\psi_k(0) = 0 \quad \Rightarrow \quad A_k + C_k = 0$$

hence,

$$C_k = -A_k$$

The slope and curvature at the base are

$$
\begin{aligned}
\psi_k'(0) &= \beta_k (B_k + D_k) \\
\psi_k''(0) &= 2 \beta_k^2 A_k
\end{aligned}
$$

Substituting into the rotational spring condition yields

$$2 EI \beta_k^2 A_k - k_\theta \beta_k (B_k + D_k) = 0$$

### 6.2 Free End $(x=L)$

The moment-free condition gives

$$A_k \left[\cosh(\beta_k L) + \cos(\beta_k L)\right] + B_k \sinh(\beta_k L) - D_k \sin(\beta_k L) = 0$$

The shear-free condition gives

$$A_k \left[\sinh(\beta_k L) - \sin(\beta_k L)\right] + B_k \cosh(\beta_k L) - D_k \cos(\beta_k L) = 0$$

---

## 7. Characteristic Equation

The three boundary equations form a homogeneous linear system

$$\mathbf{H}(\beta_k) \begin{bmatrix} A_k \\ B_k \\ D_k \end{bmatrix} = \mathbf{0}$$

Non-trivial solutions exist only when

$$\det \mathbf{H}(\beta_k) = 0$$

Solving this characteristic equation yields the admissible values of $\beta_k$, from which the natural frequencies follow as

$$\omega_k = \beta_k^2 \sqrt{\frac{EI}{\rho A}}$$

---

## 8. Mode Shapes

The corresponding mode shapes are given by

$$\psi_k(x) = A_k \cosh(\beta_k x) + B_k \sinh(\beta_k x) - A_k \cos(\beta_k x) + D_k \sin(\beta_k x)$$

where the constants are determined up to an arbitrary scaling factor.



---

## 9. Forced Vibration Using Modal Expansion

Consider a harmonic point excitation applied at $x = l_{\text{exciter}}$:

$$q(x,t) = Q \sin(\Omega t)\, \delta(x - l_{\text{exciter}})$$

The displacement response is expanded in modal coordinates as

$$y(x,t) = \sum_{k=1}^{\infty} z_k(t)\, \psi_k(x)$$

Substituting into the governing equation and applying modal orthogonality leads to the decoupled equations

$$m_k \ddot z_k + 2 \zeta_k \omega_k m_k \dot z_k + m_k \omega_k^2 z_k = Q \psi_k(l_{\text{exciter}}) \sin(\Omega t)$$

where the modal mass is

$$m_k = \int_0^L \rho A \psi_k^2(x)\, dx$$

and

$$\zeta_k = \frac{C}{2 \omega_k}$$

---

## 10. Steady-State Solution

The steady-state solution of each modal coordinate is

$$z_k(t) = \chi_k(\Omega) \cos\!\left(\Omega t + \phi_k(\Omega)\right)$$

with amplitude

$$\chi_k(\Omega) = \left| \frac{Q \psi_k(l_{\text{exciter}})} {m_k \sqrt{(\omega_k^2 - \Omega^2)^2 + 4 \zeta_k^2 \omega_k^2 \Omega^2}} \right|$$

and phase lag

$$\phi_k(\Omega) = \tan^{-1} \left( \frac{2 \zeta_k \omega_k \Omega} {\omega_k^2 - \Omega^2} \right)$$

---

## 11. Modal Damping Matrix

The damping matrix for the beam modes is diagonalized through the orthogonality of the mode shapes. For the first two dominant modes, the beam damping matrix is:

$$\mathbf{C}_{\text{beam}} = \begin{bmatrix} 2 \zeta_1 \omega_1 m_1 & 0 \\ 0 & 2 \zeta_2 \omega_2 m_2 \end{bmatrix}$$

---

## 12. Reduction to a 2-DOF Model

Retaining only the first two dominant modes,

$$y(x,t) \approx z_1(t)\psi_1(x) + z_2(t)\psi_2(x)$$

The reduced equations of motion are

$$\mathbf{M}_{\text{beam}} \ddot{\mathbf z} + \mathbf{C}_{\text{beam}} \dot{\mathbf z} + \mathbf{K}_{\text{beam}} \mathbf z = \mathbf{F}_{\text{exciter}}$$

where

$$\mathbf{M}_{\text{beam}} = \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix}, \quad \mathbf{K}_{\text{beam}} = \begin{bmatrix} m_1 \omega_1^2 & 0 \\ 0 & m_2 \omega_2^2 \end{bmatrix}$$

---

## 13. Inclusion of Localised Attachments

Using Rayleigh's approximation, the effective mass for localised attachments at position $l_i$ is determined by the kinetic energy contribution. For the spring attachment:

$$\mathbf{M}_{\text{spring}} = m_{\text{eff,s}} \begin{bmatrix} \psi_1^2(l_{\text{spring}}) & \psi_1(l_{\text{spring}})\psi_2(l_{\text{spring}}) \\ \psi_1(l_{\text{spring}})\psi_2(l_{\text{spring}}) & \psi_2^2(l_{\text{spring}}) \end{bmatrix}$$

where $m_{\text{eff,s}} = \frac{1}{3} m_{\text{spring}} + m_{\text{fixed,s}}$.

Similarly, for the localised damper and exciter with masses $m_d$ and $m_e$ at positions $l_{\text{damper}}$ and $l_{\text{exciter}}$:

$$\mathbf{M}_{\text{damper}} = m_d \begin{bmatrix} \psi_1^2(l_{\text{damper}}) & \psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) \\ \psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) & \psi_2^2(l_{\text{damper}}) \end{bmatrix}$$

$$\mathbf{M}_{\text{exciter}} = m_e \begin{bmatrix} \psi_1^2(l_{\text{exciter}}) & \psi_1(l_{\text{exciter}})\psi_2(l_{\text{exciter}}) \\ \psi_1(l_{\text{exciter}})\psi_2(l_{\text{exciter}}) & \psi_2^2(l_{\text{exciter}}) \end{bmatrix}$$

---

## 14. Localized Stiffness and Damping Matrices

For a localized damper with coefficient $c_d$ and localized spring $k_d$, the additional matrices are:

$$\mathbf{C}_{\text{damper}} = c_d \begin{bmatrix} \psi_1^2(l_{\text{damper}}) & \psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) \\ \psi_1(l_{\text{damper}})\psi_2(l_{\text{damper}}) & \psi_2^2(l_{\text{damper}}) \end{bmatrix}$$

$$\mathbf{K}_{\text{attachment}} = k_d \begin{bmatrix} \psi_1^2(l_{\text{spring}}) & \psi_1(l_{\text{spring}})\psi_2(l_{\text{spring}}) \\ \psi_1(l_{\text{spring}})\psi_2(l_{\text{spring}}) & \psi_2^2(l_{\text{spring}}) \end{bmatrix}$$

---

## 15. The Modal Force Vector

The vector $\mathbf{F}_{\text{exciter}}$ represents the projection of the external point force onto the chosen mode shapes. For a 2-DOF model, this is defined as:

$$\mathbf{F}_{\text{exciter}} = \begin{bmatrix} Q \psi_1(l_{\text{exciter}}) \\ Q \psi_2(l_{\text{exciter}}) \end{bmatrix} \sin(\Omega t)$$

---

## 16. Final Reduced System

The final 2-DOF system is

$$\left( \mathbf{M}_{\text{beam}} + \mathbf{M}_{\text{spring}} + \mathbf{M}_{\text{damper}} + \mathbf{M}_{\text{exciter}} \right) \ddot{\mathbf z} + \left( \mathbf{C}_{\text{beam}} + \mathbf{C}_{\text{damper}} \right) \dot{\mathbf z} + \left( \mathbf{K}_{\text{beam}} + \mathbf{K}_{\text{attachment}} \right) \mathbf z = \mathbf{F}_{\text{exciter}}$$

# Activity 2: Estimation of Damping Ratios of the First Two Dominant Modes

## 1. Theoretical Basis

To estimate the damping ratios $\zeta_1$ and $\zeta_2$, we analyze the discretized 2-DOF equation of motion derived in Activity 1:

$$
\mathbf{M}\ddot{\mathbf{z}} + \mathbf{C}\dot{\mathbf{z}} + \mathbf{K}\mathbf{z} = \mathbf{0}
$$

The system damping is predominantly determined by the discrete viscous damper located at $x_{damper} = 0.2$ m. To select the appropriate damping model, we examine the system's mode shapes.

## 2. Damping Model Selection

Using the theoretical model from Activity 1, we observe the mode shape amplitudes at the damper location:
* Mode 1: The damper is near a node ($|\psi_1| \approx 0$).
* Mode 2: The damper is near an antinode ($|\psi_2| \gg |\psi_1|$).

Since the damping is significantly more effective at higher frequencies, we assume Rayleigh Proportional Damping:

$$
\mathbf{C} = \alpha \mathbf{M} + \beta \mathbf{K}
$$

where $\alpha$ is the mass-proportional coefficient and $\beta$ is the stiffness-proportional coefficient.

## 3. Derivation of the Damping Ratio Formula

The modal damping ratio $\zeta_k$ for the $k^{th}$ mode is defined by projecting the damping matrix onto the mode shape vector $\phi_k$:

$$
\zeta_k = \frac{\phi_k^T \mathbf{C} \phi_k}{2\omega_k \phi_k^T \mathbf{M} \phi_k}
$$

Substitute the Rayleigh damping expression $\mathbf{C} = \alpha \mathbf{M} + \beta \mathbf{K}$ into the numerator:

$$
\phi_k^T \mathbf{C} \phi_k = \phi_k^T (\alpha \mathbf{M} + \beta \mathbf{K}) \phi_k
$$
$$
\phi_k^T \mathbf{C} \phi_k = \alpha (\phi_k^T \mathbf{M} \phi_k) + \beta (\phi_k^T \mathbf{K} \phi_k)
$$

Recall the eigenvalue property $\mathbf{K}\phi_k = \omega_k^2 \mathbf{M}\phi_k$. Multiplying by $\phi_k^T$:
$$
\phi_k^T \mathbf{K} \phi_k = \omega_k^2 (\phi_k^T \mathbf{M} \phi_k)
$$

Substituting this back into the damping equation:

$$
\phi_k^T \mathbf{C} \phi_k = (\alpha + \beta \omega_k^2) (\phi_k^T \mathbf{M} \phi_k)
$$

Now, substituting this numerator back into the definition of $\zeta_k$:

$$
\zeta_k = \frac{(\alpha + \beta \omega_k^2) (\phi_k^T \mathbf{M} \phi_k)}{2\omega_k (\phi_k^T \mathbf{M} \phi_k)}
$$

Simplifying the expression (the modal mass terms cancel out):

$$
\zeta_k = \frac{1}{2} \left( \frac{\alpha}{\omega_k} + \beta \omega_k \right)
$$

## 4. Estimation Formula for $\zeta_2$

As justified in Section 2, the damper has negligible effect on the low-frequency Mode 1 (near a node), so we assume the mass-proportional term is zero ($\alpha \approx 0$). This leaves Stiffness Proportional Damping:

$$
\zeta_k = \frac{\beta \omega_k}{2}
$$

We can now form a ratio between the second and first modes to eliminate $\beta$:

$$
\frac{\zeta_2}{\zeta_1} = \frac{\frac{\beta \omega_2}{2}}{\frac{\beta \omega_1}{2}} = \frac{\omega_2}{\omega_1}
$$

Rearranging for $\zeta_2$:

$$
\zeta_2 = \zeta_1 \left( \frac{\omega_2}{\omega_1} \right)
$$

## 5. Estimation Procedure

Since the exciter's maximum operating frequency is typically below the second natural frequency ($\omega_2$), the second damping peak cannot be measured experimentally. The following procedure is used to estimate $\zeta_2$:

1.  Measure $\zeta_1$: Obtain the frequency response curve experimentally. Use the Half-Power Bandwidth Method on the first resonance peak to calculate the experimental value of $\zeta_1$.
2.  Compute Natural Frequencies: Use the theoretical model derived in Activity 1 to calculate the natural frequencies $\omega_1$ and $\omega_2$.
3.  Calculate $\zeta_2$: Substitute the measured $\zeta_1$ and the calculated frequencies into the derived formula:

$$
\zeta_{2, estimated} = \zeta_{1, measured} \times \frac{\omega_2}{\omega_1}
$$
