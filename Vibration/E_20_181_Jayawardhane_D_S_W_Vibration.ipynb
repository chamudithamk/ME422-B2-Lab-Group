## Activity 01
# Stage A: Beam only (Infinite DOF → 2-DOF modal model)

## 1) Infinite DOF beam PDE (Euler–Bernoulli)
For small transverse vibration:

$$
\rho A \, \frac{\partial^2 y}{\partial t^2}(x,t) + EI \, \frac{\partial^4 y}{\partial x^4}(x,t)=f(x,t)
$$

For the *beam-only free vibration* (no external force):

$$
\rho A \, y_{tt} + EI \, y_{xxxx}=0
$$

This is infinite DOF because \(y\) depends on continuous position \(x in[0,L]\).

---

## 2) Mode shapes (eigenfunctions)
Assume separable motion:

$$
y(x,t)=\psi(x)\,q(t)
$$

Substitute into the free-vibration PDE:

$$
\rho A \,\psi(x)\,\ddot q(t) + EI\,\psi''''(x)\,q(t)=0
$$

Divide by $(\psi q)$:

$$
\frac{\psi''''}{\psi} = -\frac{\rho A}{EI}\frac{\ddot q}{q}=\beta^4
$$

So:

### Spatial ODE (eigenvalue problem)
$$
\psi''''(x)-\beta^4\psi(x)=0
$$

### Time ODE
$$
\ddot q(t)+\omega^2 q(t)=0
$$

with the relationship

$$
\omega=\beta^2\sqrt{\frac{EI}{\rho A}}
$$

Applying boundary conditions gives discrete roots $(\beta_k)$ and mode shapes $(\psi_k(x))$.

---

## 3) Infinite modal expansion
Any displacement can be written as an infinite sum:

$$
y(x,t)=\sum_{k=1}^{\infty} z_k(t)\,\psi_k(x)
$$

This still has infinitely many coordinates $(z_1,z_2,z_3,$...).

---

## 4) Truncate to first two dominant modes (2 DOF)
Approximate using only first two modes:

$$
y(x,t)\approx z_1(t)\psi_1(x)+z_2(t)\psi_2(x)
$$

Define the 2-DOF coordinate vector:

$$
\mathbf z(t)=
\begin{bmatrix}
z_1(t)\\
z_2(t)
\end{bmatrix}
$$

---

## 5) Galerkin projection → 2×2 equations
Start from the forced PDE again:

$$
\rho A\,y_{tt}+EI\,y_{xxxx}=f(x,t)
$$

Insert the 2-mode approximation. Multiply both sides by $(\psi_i(x))$ and integrate over \(0\) to \(L\) for \(i=1,2\):

$$
\int_0^L \psi_i \left(\rho A\,y_{tt}+EI\,y_{xxxx}\right)\,dx = \int_0^L \psi_i f(x,t)\,dx
$$

This yields:

$$
\sum_{j=1}^{2} M_{ij}\ddot z_j + \sum_{j=1}^{2} K_{ij} z_j = Q_i(t)
$$

where

### Mass matrix
$$
M_{ij}=\int_0^L \rho A\,\psi_i(x)\psi_j(x)\,dx
$$

### Stiffness matrix
(standard symmetric form)

$$
K_{ij}=\int_0^L EI\,\psi_i''(x)\psi_j''(x)\,dx
$$

### Generalized force
$$
Q_i(t)=\int_0^L \psi_i(x)\,f(x,t)\,dx
$$

---

## 6) Beam-only 2-DOF model (matrix form)
$$
\boxed{
\mathbf M_{\text{beam}}\ddot{\mathbf z}+\mathbf K_{\text{beam}}\mathbf z = \mathbf Q(t)
}
$$

If $(\psi_1,\psi_2)$ are true beam modes, orthogonality makes the beam matrices diagonal:

$$
\mathbf M_{\text{beam}}=
\begin{bmatrix}
m_1 & 0\\
0 & m_2
\end{bmatrix}
$$

and

$$
\mathbf K_{\text{beam}}=
\begin{bmatrix}
\omega_1^2 m_1 & 0\\
0 & \omega_2^2 m_2
\end{bmatrix}
$$

If modal damping ratios \(\zeta_1,\zeta_2\) are assumed:

$$
\mathbf C_{\text{beam}}=
\begin{bmatrix}
2\zeta_1\omega_1 m_1 & 0\\
0 & 2\zeta_2\omega_2 m_2
\end{bmatrix}
$$

So the beam-only 2-DOF form is:

$$
\boxed{
\mathbf M_{\text{beam}}\ddot{\mathbf z}
+\mathbf C_{\text{beam}}\dot{\mathbf z}
+\mathbf K_{\text{beam}}\mathbf z
=
\mathbf F_{\text{modal}}(t)
}
$$


# Stage B: 2-DOF reduced model for the system with damper

## B1) Two-mode approximation
$$
y(x,t)\approx z_1(t)\,\psi_{1}(x)+z_2(t)\,\psi_{2}(x),
\qquad
\mathbf{z}(t)=
\begin{bmatrix}
z_1(t)\\
z_2(t)
\end{bmatrix}.
$$

Define the mode-shape vector:
$$
\boldsymbol{\psi}(x)=
\begin{bmatrix}
\psi_{1}(x)\\
\psi_{2}(x)
\end{bmatrix}.
$$

---

## B2) Beam-only modal matrices (Stage A result)
Beam-only (modal) matrices are diagonal:
$$
\mathbf{M}_{\text{beam}}=
\begin{bmatrix}
m_1 & 0\\
0 & m_2
\end{bmatrix},
\qquad
\mathbf{K}_{\text{beam}}=
\begin{bmatrix}
\omega_1^{2}m_1 & 0\\
0 & \omega_2^{2}m_2
\end{bmatrix}.
$$

Modal damping for the beam (if assumed):
$$
\mathbf{C}_{\text{beam}}=
\begin{bmatrix}
2\zeta_1\omega_1 m_1 & 0\\
0 & 2\zeta_2\omega_2 m_2
\end{bmatrix}.
$$

---

## B3) Kinematics at any attachment point
At a point \(x=a\):
$$
y(a,t)=\boldsymbol{\psi}(a)^{T}\mathbf{z}(t),
\qquad
\dot{y}(a,t)=\boldsymbol{\psi}(a)^{T}\dot{\mathbf{z}}(t),
$$
where
$$
\boldsymbol{\psi}(a)=
\begin{bmatrix}
\psi_{1}(a)\\
\psi_{2}(a)
\end{bmatrix}.
$$

---

## B4) Damper mass at $(x=l_{\text{damper}})$
The lumped mass $m_{\text{damper}}$ contributes:
$$
\mathbf{M}_{\text{damper}}
=
m_{\text{damper}}\,
\boldsymbol{\psi}(l_{\text{damper}})
\boldsymbol{\psi}(l_{\text{damper}})^{T}.
$$

---

## B5) Exciter mass at $(x=l_{exciter})$
The lumped mass $m_{\text{exciter}}$ contributes:
$$
\mathbf{M}_{\text{exciter}}
=
m_{\text{exciter}}\,
\boldsymbol{\psi}(l_{\text{exciter}})
\boldsymbol{\psi}(l_{\text{exciter}})^{T}.
$$

---

## B6) Dashpot damper coefficient $(c_d$) at $x=l_{damper}$
The physical damper to ground adds damping:
$$
\mathbf{C}_{\text{damper}}
=
c_d\,
\boldsymbol{\psi}(l_{\text{damper}})
\boldsymbol{\psi}(l_{\text{damper}})^{T}.
$$

---

## B7) Tip spring stiffness $k_s$ at $x=L$
The spring to ground at the tip adds stiffness:
$$
\mathbf{K}_{\text{spring}}
=
k_s\,
\boldsymbol{\psi}(L)\boldsymbol{\psi}(L)^{T}.
$$

---

## B8) Final reduced 2-DOF equation
Total matrices:
$$
\mathbf{M}=\mathbf{M}_{\text{beam}}+\mathbf{M}_{\text{damper}}+\mathbf{M}_{\text{exciter}},
$$
$$
\mathbf{C}=\mathbf{C}_{\text{beam}}+\mathbf{C}_{\text{damper}},
$$
$$
\mathbf{K}=\mathbf{K}_{\text{beam}}+\mathbf{K}_{\text{spring}}.
$$

Final equation:
$$
\boxed{
\mathbf{M}\ddot{\mathbf{z}}+\mathbf{C}\dot{\mathbf{z}}+\mathbf{K}\mathbf{z}
=
\mathbf{F}_{\text{exciter}}
}
$$

Equivalently:
$$
\boxed{
\left(\mathbf{M}_{\text{beam}}+\mathbf{M}_{\text{damper}}+\mathbf{M}_{\text{exciter}}\right)\ddot{\mathbf{z}}
+
\left(\mathbf{C}_{\text{beam}}+\mathbf{C}_{\text{damper}}\right)\dot{\mathbf{z}}
+
\left(\mathbf{K}_{\text{beam}}+\mathbf{K}_{\text{spring}}\right)\mathbf{z}
=
\mathbf{F}_{\text{exciter}}
}
$$


# Activity 2: Estimation of Damping Ratios of the First Two Dominant Modes



## Starting point (reduced system)
From Activity 1, the reduced equations of motion are

$$
\mathbf{M}\ddot{\mathbf{z}} + \mathbf{C}\dot{\mathbf{z}} + \mathbf{K}\mathbf{z} = \mathbf{0}
$$

where
$$
\mathbf{z}(t)=
\begin{bmatrix}
z_1(t)\\
z_2(t)
\end{bmatrix}.
$$

---

## Assumption: Rayleigh proportional damping
The damping matrix is assumed to be proportional to the mass and stiffness matrices:

$$
\boxed{
\mathbf{C} = \alpha\,\mathbf{M} + \beta\,\mathbf{K}
}
$$

where  
- $\alpha$ is the mass-proportional damping coefficient  
- $\beta\$ is the stiffness-proportional damping coefficient  

This assumption preserves modal orthogonality.

---

## Modal damping ratio definition
For the \(k\)-th mode, the damping ratio is defined as

$$
\zeta_k =
\frac{\boldsymbol{\phi}_k^{T}\mathbf{C}\boldsymbol{\phi}_k}
{2\omega_k\,\boldsymbol{\phi}_k^{T}\mathbf{M}\boldsymbol{\phi}_k}
$$

where  
- $\boldsymbol{\phi}_k$ is the mode shape vector  
- $\omega_k$ is the natural frequency of the \(k\)-th mode  

---

## Substitution of Rayleigh damping
Substituting $\mathbf{C}=\alpha\mathbf{M}+\beta\mathbf{K}$:

$$
\boldsymbol{\phi}_k^{T}\mathbf{C}\boldsymbol{\phi}_k
=
\alpha\,\boldsymbol{\phi}_k^{T}\mathbf{M}\boldsymbol{\phi}_k
+
\beta\,\boldsymbol{\phi}_k^{T}\mathbf{K}\boldsymbol{\phi}_k
$$

Using the eigenvalue relation:

$$
\mathbf{K}\boldsymbol{\phi}_k = \omega_k^2 \mathbf{M}\boldsymbol{\phi}_k
$$

we obtain:

$$
\boldsymbol{\phi}_k^{T}\mathbf{K}\boldsymbol{\phi}_k
=
\omega_k^2 \boldsymbol{\phi}_k^{T}\mathbf{M}\boldsymbol{\phi}_k
$$

---

## Final damping ratio expression
Therefore, the damping ratio of the \(k\)-th mode is

$$
\boxed{
\zeta_k =
\frac{1}{2}
\left(
\frac{\alpha}{\omega_k}
+
\beta\,\omega_k
\right)
}
$$

---

## Damping ratios of the first two modes
Using the natural frequencies $\omega_1$ and $\omega_2$:

$$
\zeta_1 =
\frac{1}{2}
\left(
\frac{\alpha}{\omega_1}
+
\beta\,\omega_1
\right),
\qquad
\zeta_2 =
\frac{1}{2}
\left(
\frac{\alpha}{\omega_2}
+
\beta\,\omega_2
\right)
$$

These values represent the estimated damping ratios of the first two dominant modes.


```python
import numpy as np
from scipy.integrate import quad

# ============================================================
# PARAMETERS (FILL THESE with your lab values)
# ============================================================
L   = 0.50          # beam length (m)
rho = 7800.0        # density (kg/m^3)
A   = 2.0e-4        # cross-sectional area (m^2)
E   = 2.0e11        # Young's modulus (Pa)
I   = 1.0e-10       # second moment of area (m^4)
EI  = E*I

# Attachment locations (m)
l_damper  = 0.30    # location of damper + damper mass (m)
l_exciter = 0.45    # location of exciter mass + force (m)

# Attachment values
m_damper  = 0.20    # damper lumped mass (kg)
m_exciter = 0.30    # exciter lumped mass (kg)
c_d       = 10.0    # damper coefficient (N·s/m)
k_s       = 500.0   # tip spring stiffness (N/m)

# Rayleigh damping coefficients (Activity 2 method)
alpha = 0.01        # mass-proportional coefficient
beta  = 1e-5        # stiffness-proportional coefficient

# ============================================================
# MODE SHAPES for CANTILEVER (clamped-free)
# Using standard roots beta*L = 1.875..., 4.694...
# ============================================================
betaL1 = 1.875104068711961
betaL2 = 4.694091132974174
beta1 = betaL1 / L
beta2 = betaL2 / L

def alpha_mode(beta):
    return (np.cosh(beta*L) + np.cos(beta*L)) / (np.sinh(beta*L) + np.sin(beta*L))

def psi(x, beta):
    a = alpha_mode(beta)
    return (np.cosh(beta*x) - np.cos(beta*x)) - a*(np.sinh(beta*x) - np.sin(beta*x))

# ============================================================
# ACTIVITY 1: Build beam modal quantities (m1, m2, omega1, omega2)
# ============================================================
def modal_mass(beta_i, beta_j):
    f = lambda x: rho*A*psi(x, beta_i)*psi(x, beta_j)
    return quad(f, 0, L)[0]

# Beam modal mass matrix from integrals
M_beam = np.array([[modal_mass(beta1, beta1), modal_mass(beta1, beta2)],
                   [modal_mass(beta2, beta1), modal_mass(beta2, beta2)]], dtype=float)

# Beam natural frequencies (Euler–Bernoulli relation)
omega1_beam = (beta1**2)*np.sqrt(EI/(rho*A))
omega2_beam = (beta2**2)*np.sqrt(EI/(rho*A))

# Beam stiffness in modal coordinates (standard approach used in your file)
K_beam = M_beam @ np.diag([omega1_beam**2, omega2_beam**2])

# Helper: attachment shape vector at x=a
def pvec(a):
    return np.array([psi(a, beta1), psi(a, beta2)], dtype=float).reshape(2,1)

# ============================================================
# ACTIVITY 1: Add attachments for YOUR system
# System: damper mass at l_damper, exciter mass at l_exciter,
# damper c_d at l_damper, spring k_s at x=L
# ============================================================
M = M_beam.copy()
C = np.zeros((2,2), dtype=float)
K = K_beam.copy()

# Add damper lumped mass
p_d = pvec(l_damper)
M += m_damper * (p_d @ p_d.T)

# Add exciter lumped mass
p_e = pvec(l_exciter)
M += m_exciter * (p_e @ p_e.T)

# Add physical damper to ground at l_damper
C += c_d * (p_d @ p_d.T)

# Add tip spring to ground at x=L
p_L = pvec(L)
K += k_s * (p_L @ p_L.T)

# ============================================================
# Natural frequencies of the TOTAL (attached) 2-DOF system
# Solve (K - ω^2 M)φ = 0
# ============================================================
eigvals = np.linalg.eigvals(np.linalg.inv(M) @ K)
eigvals = np.real(eigvals)
eigvals.sort()

omega1 = np.sqrt(eigvals[0])
omega2 = np.sqrt(eigvals[1])

print("=== ACTIVITY 1 RESULTS (with attachments) ===")
print("M (total) =\n", M)
print("C (from damper) =\n", C)
print("K (total) =\n", K)
print("\nNatural frequencies (rad/s):")
print("omega1 =", omega1)
print("omega2 =", omega2)

# ============================================================
# ACTIVITY 2: Damping ratios using Rayleigh method (from your file)
# ζ_k = 1/2 ( α/ω_k + β ω_k )
# ============================================================
zeta1 = 0.5 * (alpha/omega1 + beta*omega1)
zeta2 = 0.5 * (alpha/omega2 + beta*omega2)

print("\n=== ACTIVITY 2 RESULTS (Rayleigh damping ratios) ===")
print("alpha =", alpha, " beta =", beta)
print("zeta1 =", zeta1)
print("zeta2 =", zeta2)


```

# Activity 3: System responses using the 2-mode (2-DOF) model

We use the reduced model from Activity 1:

$$
\mathbf{M}\ddot{\mathbf{z}} + \mathbf{C}\dot{\mathbf{z}} + \mathbf{K}\mathbf{z} = \mathbf{F}(t),
\qquad
\mathbf{z}(t)=\begin{bmatrix}z_1\\z_2\end{bmatrix}.
$$

## Modal transformation of the attached system
Let $\Phi be the eigenvector matrix of the attached system from:

$$
(\mathbf{K}-\omega^2\mathbf{M})\boldsymbol{\phi}=\mathbf{0}.
$$

Define modal coordinates $\mathbf{q}(t)$ such that:

$$
\mathbf{z}(t)=\Phi\,\mathbf{q}(t).
$$

The displacement can be written using **modified mode shapes**:

$$
y(x,t)=\psi_1^*(x)\,q_1(t)+\psi_2^*(x)\,q_2(t),
$$

where

$$
\psi_1^*(x)=\phi_{11}\psi_1(x)+\phi_{21}\psi_2(x),
\qquad
\psi_2^*(x)=\phi_{12}\psi_1(x)+\phi_{22}\psi_2(x).
$$

---

## 3.1 Free vibration (unforced)
No external force: $\mathbf{F}(t)=\mathbf{0}\$.  
We impose an initial displacement at the beam end \(x=L\):

$$
y(L,0)=y_0,\qquad \dot{y}(L,0)=0.
$$

We choose $q_2(0)=0$ and compute:

$$
q_1(0)=\frac{y_0}{\psi_1^*(L)}.
$$

---

## 3.2 Forced vibration at three motor speeds
A harmonic force acts at the exciter location $x=l_{\text{exciter}}$:

$$
F(t)=Q(\Omega)\sin(\Omega t).
$$

The modal forcing vector becomes:

$$
\mathbf{F}_q(t)=
\begin{bmatrix}
\psi_1^*(l_{\text{exciter}})\\
\psi_2^*(l_{\text{exciter}})
\end{bmatrix}
Q(\Omega)\sin(\Omega t).
$$

---

## 3.3 Frequency response function (FRF)
For each position \(x\), the FRF is:

$$
G(x,\Omega)=\frac{Y(x,\Omega)}{Q(\Omega)}.
$$

Using the two modified modes, the complex steady-state response is computed and
magnitude and phase are plotted for selected beam locations.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# -----------------------------
# REQUIRED inputs from Activity 1/2
# -----------------------------
# M, C, K : 2x2 total matrices (attachments included)
# beta1, beta2 and psi(x,beta) : bare beam mode shapes
# L : beam length
# l_exciter, l_damper : locations

# ---- quick sanity check ----
for name in ["M","K","C","L","l_exciter","l_damper","beta1","beta2","psi"]:
    if name not in globals():
        raise NameError(f"Missing '{name}'. Run Activity 1 cells first or define {name}.")

M = np.array(M, dtype=float)
C = np.array(C, dtype=float)
K = np.array(K, dtype=float)

# -----------------------------
# Eigen-solution of attached system
# -----------------------------
A = np.linalg.inv(M) @ K
eigvals, eigvecs = np.linalg.eig(A)
eigvals = np.real(eigvals)

idx = np.argsort(eigvals)
eigvals = eigvals[idx]
Phi = np.real(eigvecs[:, idx])

omega_n = np.sqrt(eigvals)  # [omega1, omega2]

omega1 = float(omega_n[0])
omega2 = float(omega_n[1])

print("omega1 (rad/s) =", omega1)
print("omega2 (rad/s) =", omega2)
print("Phi =\n", Phi)

# -----------------------------
# Modified (star) mode shapes
# psi_star_1(x) = phi11*psi1(x) + phi21*psi2(x)
# psi_star_2(x) = phi12*psi1(x) + phi22*psi2(x)
# -----------------------------
def psi1(x): return psi(x, beta1)
def psi2(x): return psi(x, beta2)

def psi_star_1(x):
    return Phi[0,0]*psi1(x) + Phi[1,0]*psi2(x)

def psi_star_2(x):
    return Phi[0,1]*psi1(x) + Phi[1,1]*psi2(x)

# vector form
def psi_star_vec(x):
    return np.array([psi_star_1(x), psi_star_2(x)], dtype=float).reshape(2,1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ------------------------------------------------
# Activity 3.1: Unforced response (gradual decay)
# ------------------------------------------------

# ---- Damping (make decay slow) ----
zeta1 = 0.01          # 1% damping
beta_rayleigh = 2*zeta1/omega1
C = beta_rayleigh * K   # stiffness-proportional damping

# ---- Initial displacement at free end ----
y0 = 0.003   # meters (same scale as your plot)

# Convert to modal coordinates (excite mainly 1st mode)
z10 = y0 / psi1(L)
z20 = 0.0

dz10 = 0.0
dz20 = 0.0

X0 = np.array([z10, z20, dz10, dz20])

# ---- Free vibration ODE ----
def free_response(t, X):
    z = X[0:2].reshape(2,1)
    dz = X[2:4].reshape(2,1)
    ddz = np.linalg.inv(M) @ (-C @ dz - K @ z)
    return np.hstack([dz.flatten(), ddz.flatten()])

# ---- Time integration ----
t_end = 10.0
t_eval = np.linspace(0, t_end, 6000)

sol = solve_ivp(
    free_response,
    (0, t_end),
    X0,
    t_eval=t_eval,
    rtol=1e-8,
    atol=1e-10
)

# ---- Free-end displacement ----
z_t = sol.y[0:2, :]
yL = psi1(L)*z_t[0, :] + psi2(L)*z_t[1, :]

# ---- Plot ----
plt.figure(figsize=(8,4))
plt.plot(sol.t, yL)
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Unforced Response at the Free End of the Beam")
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from matplotlib import animation
from IPython.display import HTML, display

# -----------------------------
# REQUIRED checks
# -----------------------------
for name in ["M","C","K","L","l_exciter","beta1","beta2","psi"]:
    if name not in globals():
        raise NameError(f"Missing '{name}'. Run Activity 1 cells first or define {name}.")

M = np.array(M, dtype=float)
C = np.array(C, dtype=float)
K = np.array(K, dtype=float)

def psi1(x): return psi(x, beta1)
def psi2(x): return psi(x, beta2)

# -----------------------------
# Motor speeds (rad/s) from file
# -----------------------------
Omega_list = [32.472, 100.0, 204.048]

# -----------------------------
# Excitation force model (edit to match your file)
# Q = 2*m_ecc*r*Omega^2
# -----------------------------
m_ecc = 0.01  # kg (EDIT)
r     = 0.01  # m  (EDIT)

def Q_of_Omega(Omega):
    return 2*m_ecc*r*(Omega**2)

# Force direction vector in z coordinates (point force at x = l_exciter)
p_exc = np.array([psi1(l_exciter), psi2(l_exciter)], dtype=float).reshape(2,1)

def forced_ode_factory(Omega):
    Qamp = Q_of_Omega(Omega)

    def ode(t, X):
        z  = X[0:2].reshape(2,1)
        dz = X[2:4].reshape(2,1)
        f = p_exc * (Qamp*np.sin(Omega*t))
        ddz = np.linalg.inv(M) @ (f - C@dz - K@z)
        return np.hstack([dz.flatten(), ddz.flatten()])

    return ode

# Beam points for animation
x_grid = np.linspace(0, L, 200)

def reconstruct_y(z_vec):
    return psi1(x_grid)*z_vec[0] + psi2(x_grid)*z_vec[1]

# -----------------------------
# Animation function
# -----------------------------
def animate_forced_response(Omega, t_end=6.0, n_steps=1200, settle_ratio=0.3, target_amp=0.005):

    t_eval = np.linspace(0, t_end, n_steps)

    # Start from rest
    X0 = np.zeros(4)

    sol = solve_ivp(
        forced_ode_factory(Omega),
        (0, t_end),
        X0,
        t_eval=t_eval,
        rtol=1e-7,
        atol=1e-9
    )

    z_t = sol.y[0:2, :].T  # (N,2)

    # Skip early transients so animation shows steady-state motion
    start_idx = int(settle_ratio * len(t_eval))
    z_anim = z_t[start_idx:, :]

    # Precompute y(x,t) frames
    Y_frames = np.array([reconstruct_y(z_anim[i,:]) for i in range(len(z_anim))])

    # Plot setup
    fig, ax = plt.subplots(figsize=(7, 3.5))
    line, = ax.plot([], [], lw=2)

    ax.set_xlim(0, L)

    # Set y-limits based on normalized target amplitude
    ax.set_ylim(-1.2*target_amp, 1.2*target_amp)

    ax.set_xlabel("Position along beam (m)")
    ax.set_ylabel("Displacement (m)")
    ax.set_title(f"Forced Response Animation, Ω = {Omega:.3f} rad/s")
    ax.grid(True)

    def init():
        line.set_data([], [])
        return (line,)

    def update(frame):
        y = Y_frames[frame]

        # normalize for visualization
        max_amp = np.max(np.abs(y))
        if max_amp > 0:
            y = (y / max_amp) * target_amp

        line.set_data(x_grid, y)
        return (line,)

    ani = animation.FuncAnimation(
        fig,
        update,
        frames=range(0,len(Y_frames),4),
        init_func=init,
        interval=20,
        blit=True
    )

    plt.close(fig)
    return ani

# -----------------------------
# Generate and display animations
# -----------------------------
for Omega in Omega_list:
    ani = animate_forced_response(Omega, t_end=6.0, n_steps=1500, settle_ratio=0.4, target_amp=0.005)
    display(HTML(ani.to_jshtml()))
```
