Derivation of the 2-DOF Model1.1 The Governing EquationThe small-amplitude transverse vibration of a slender beam is governed by the Euler-Bernoulli partial differential equation:$$\rho A \frac{\partial^2 y}{\partial t^2} + EI \frac{\partial^4 y}{\partial x^4} = f(t,x)$$1.2 Modal TruncationWe approximate the solution using the first two dominant mode shapes $\psi_1(x)$ and $\psi_2(x)$:$$y(x,t) \approx \psi_1(x)z_1(t) + \psi_2(x)z_2(t)$$where $z_1(t)$ and $z_2(t)$ are the modal coordinates.1.3 Matrix Form of the Equation of MotionUsing the principle of virtual work or Galerkin's method, the system is reduced to a 2-DOF matrix equation:$$\mathbf{M}\mathbf{\ddot{z}} + \mathbf{C}\mathbf{\dot{z}} + \mathbf{K}\mathbf{z} = \mathbf{F}(t)$$The total Mass Matrix $\mathbf{M}$ is the sum of the beam's modal mass and the contributions from attached components (Spring, Damper, Exciter):$$\mathbf{M} = \begin{bmatrix} m_1 & 0 \\ 0 & m_2 \end{bmatrix} + \sum_{p} m_p \begin{bmatrix} \psi_1^2(x_p) & \psi_1(x_p)\psi_2(x_p) \\ \psi_1(x_p)\psi_2(x_p) & \psi_2^2(x_p) \end{bmatrix}$$The Stiffness Matrix $\mathbf{K}$ is defined by the natural frequencies $\omega_n$:$$\mathbf{K} = \begin{bmatrix} m_1\omega_1^2 & 0 \\ 0 & m_2\omega_2^2 \end{bmatrix}$$
